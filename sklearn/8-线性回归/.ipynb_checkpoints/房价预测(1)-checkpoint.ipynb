```python
from scipy.stats import norm, skew
from scipy import stats
import seaborn as sns

plt.rcParams['axes.unicode_minus']=False 
sns.distplot(df['tprice'], fit=norm)

(mu, sigma) = norm.fit(df['tprice'])
print('\n mu={:.2f} and sigma={:.2f} \n'.format(mu, sigma))

plt.legend(['Normal dist.($\mu=$ {:.2f} and $\sigma=$ {:.2f})'.format(mu, sigma)], loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

fig = plt.figure()
stats.probplot(df['tprice'], plot=plt)
plt.show()
```

```python
y = df['tprice']
del df['tprice']


model = MinMaxScaler()

new_df = model.fit_transform(df)

std = StandardScaler()

new_df = std.fit_transform(df)

data_tr,data_te,label_tr,label_te = train_test_split(new_df,y,random_state=32,test_size=0.2)

model = LinearRegression()

model.fit(data_tr,label_tr)

pre = model.predict(data_te)
```

```python
error_sum = 0
for i,j in zip(label_te.tolist(),pre.tolist()):
    error_sum+=sqrt((i-j)**2)
pre_sum=sum(pre)
print(1-(error_sum/pre_sum))
```

```python
fig = plt.figure(figsize=(50,10))
plt.plot(pd.Series(pre).index,pre,color='r',label='预测房价')
plt.plot(pd.Series(pre).index,label_te,color='b',label='实际房价')
plt.xticks(pd.Series(pre).index)
plt.legend()
plt.show()
```
