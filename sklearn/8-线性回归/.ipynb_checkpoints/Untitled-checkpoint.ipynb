```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn import model_selection
from sklearn import tree
from sklearn.linear_model import  LinearRegression
from sklearn import ensemble
```

```python
df = pd.read_csv('./datas/result.csv',index_col=0)
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
del df['commName']
```

```python
df.head()
```

```python
an = np.unique(df['areaName'].values)
ri = np.unique(df['rmInfo'].values)
ele = np.unique(df['elevator'].values)
sw = np.unique(df['subway'].values)
fl = np.unique(df['floor_level'].values)
ht = np.unique(df['house_type'].values)
de = np.unique(df['decoration'].values)
tm = np.unique(df['tmInfo'].values)
ye = np.unique(df['year'].values)
```

```python
temp = map(lambda x:{i:j for i,j in zip(x,range(0,len(x)))},[an,ri,ele,sw,fl,ht,de,tm,ye])
```

```python
for i in temp:
    df = df.replace(i)
```

```python
df.head()
```

```python
data = df.loc[:,df.columns != 'uprice'].values
target = df.loc[:,df.columns == 'uprice'].values
```

```python
Xtrain, Xtest, Ytrain, Ytest = model_selection.train_test_split(data,target,test_size=0.3)
```

```python
regressor = ensemble.RandomForestRegressor(n_estimators=100,random_state=0)
model_selection.cross_val_score(regressor, data, target, cv=10
               ,scoring = "neg_mean_squared_error")
```

```python
regressor.fit(Xtrain, Ytrain)
```

```python
regressor.score(Xtest,Ytest)
```
